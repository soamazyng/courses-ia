```python
from dotenv import load_dotenv
_ = load_dotenv()

from crewai import Agent, Task, Crew, Process
```

```python
planejador_de_viagem = Agent(
    role="Planejador de Viagem",
    goal="Planejar todos os detalhes de uma viagem incluindo voos, hospedagem, roteiros, atividades e etc",
    backstory="""
Você é um especialista em viagens e sempre busca as melhores opções para seus clientes.
Você é muito organizado e sempre busca as melhores opções para seus clientes.
Você é muito profissional e sempre busca as melhores opções para seus clientes.
Você é muito responsável e sempre busca as melhores opções para seus clientes.
Você é muito confiável e sempre busca as melhores opções para seus clientes.
Você é muito eficiente e sempre busca as melhores opções para seus clientes.
Você é muito eficaz e sempre busca as melhores opções para seus clientes.
Sempre está em busta de novas aventuras e experiências.
Seu objetivo é garantir que os detalhes da viagem sejam organizados de maneira eficiente e agradável.
    """,
    verbose=True,
)
```

```python
orcamento_de_viagem = Agent(
    role="Orcamento de Viagem",
    goal="Estimar o custo total de uma viagem considerando, transporte, alimentação, atividades, hospedagem, etc",
    backstory="""
    Você é um analista financeiro focado em viagens e sua missão é garantir que os custos estejam dentro do orçamento criando estimativas precisas para cara parte da viagem.
    """,
    verbose=True,
)
```

```python
planeja_roteiro = Task(
    description="""
    Crie um roteiro detalhado para uma viagem na Europa, incluindo as cidades, atividades e transporte.
    """,
    agent=planejador_de_viagem,
    expected_output="""
    Um roteiro com a sequência de cidades a serem visitadas, as atividades a serem realizadas e o transporte a ser utilizado.
    """
)
```

```python
estima_orcamento = Task(
    description="""
    Crie um orçamento detalhado para uma viagem, incluindo as cidades, atividades e transporte.
    """,
    agent=orcamento_de_viagem,
    expected_output="""
    Um orçamento detalhado com o custo total da viagem, incluindo as cidades, atividades e transporte.
    Essa estimativa deve ser convertida para o formato de moeda do Brasil.
    """
)
```

```python
viagem_crew = Crew(
    agents=[planejador_de_viagem, orcamento_de_viagem],
    tasks=[planeja_roteiro, estima_orcamento],
    process=Process.sequential,
)
```

```python
result = viagem_crew.kickoff()
```
